```python
import math

def sin_Taylor(x, threshold=1e-4):
    y = 0
    for i in range(100000):
        c = (-1)**i / math.factorial(2*i + 1)
        t = x**(2*i + 1) * c
        y = y + t
        if abs(t.data) < threshold:
            break

    return y
```
