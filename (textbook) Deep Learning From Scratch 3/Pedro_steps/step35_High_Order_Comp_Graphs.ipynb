```python
class Tanh(Pedro_Function):
    def forward(self, x):
        y = np.tanh(x)
        return y

    def backward(self, gy):
        y = self.output[0]()
        gx = gy * (1 - y**2)
        return gx

def tanh(x):
    f = Tanh()
    return f(x)
```

```python
x = Pedro_Variable(np.array(1.0))
y = tanh(x)
x.name = 'x'
y.name = 'y'
y.backward(create_graph=True)

iters = 0

for i in range(iters):
    gx = x.grad
    x.cleargrad()
    gx.backward(create_graph=True)

gx = x.grad
gx.name = 'gx' + str(iters+1)
plot_dot_graph(gx, verbose=True, to_file='tanh.png')
```
