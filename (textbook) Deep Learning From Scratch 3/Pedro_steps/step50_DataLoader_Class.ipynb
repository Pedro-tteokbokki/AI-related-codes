```python
from Pedro_package import no_grad

max_epoch = 300
batch_size = 30
hidden_size = 10
lr = 1.0

train_set = datasets.Spiral(train=True)
test_set = datasets.Spiral(train=False)
train_loader = dataloaders.DataLoader(train_set, batch_size=batch_size)
test_loader = dataloaders.DataLoader(test_set, batch_size=batch_size, shuffle=False)

model = Pedro_MLP((hidden_size, 3))
optimizer = optimizers.SGD(lr).setup(model)
sum_loss_train = np.zeros(max_epoch)
sum_acc_train = np.zeros(max_epoch)
sum_loss_test = np.zeros(max_epoch)
sum_acc_test = np.zeros(max_epoch)

for epoch in range(max_epoch):

    for x, t in train_loader:
        y = model(x)
        loss = F.softmax_cross_entropy(y, t)
        acc = F.accuracy(y, t)
        model.cleargrads()
        loss.backward()
        optimizer.update()

        sum_loss_train[epoch] += float(loss.data) * len(t)
        sum_acc_train[epoch] += float(acc.data) * len(t)

    print('epoch: {}'.format(epoch+1))
    print('train loss: {:.4f}, accuracy: {:.4f}'.format(
        sum_loss_train[epoch] / len(train_set), sum_acc_train[epoch] / len(train_set)))

    sum_loss, sum_acc = 0, 0
    with no_grad():
        for x, t in test_loader:
            y = model(x)
            loss = F.softmax_cross_entropy(y, t)
            acc = F.accuracy(y, t)
            sum_loss_test[epoch] += float(loss.data) * len(t)
            sum_acc_test[epoch] += float(acc.data) * len(t)

    print('test loss: {:.4f}, accuracy: {:.4f}'.format(
        sum_loss_test[epoch] / len(test_set), sum_acc_test[epoch] / len(test_set)))


plt.subplot(121)
plt.plot(np.arange(max_epoch), sum_loss_train / len(train_set), 'b', label='train')
plt.plot(np.arange(max_epoch), sum_loss_test / len(test_set), 'r', label='test')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()

plt.subplot(122)
plt.plot(np.arange(max_epoch), sum_acc_train / len(train_set), 'b', label='train')
plt.plot(np.arange(max_epoch), sum_acc_test / len(test_set), 'r', label='test')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend()

plt.show()
```
