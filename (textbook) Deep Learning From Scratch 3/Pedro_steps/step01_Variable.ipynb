# Defining simple variable

```python
class Pedro_Variable:
    def __init__(self, data):
        self.data = data
```

```python
data = np.array(1.0)
x = Pedro_Variable(data)
print(type(x))
print(x.data)
```

```python
x.data = np.array(2.0)
print(x.data)
```
