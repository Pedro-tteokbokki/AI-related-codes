```python
def rosenbrock(x0, x1):
    y = 100 * (x1 - x0**2)**2 + (1 - x0)**2
    return y
```

```python
x0 = Pedro_Variable(np.array(0.0))
x1 = Pedro_Variable(np.array(2.0))
y = rosenbrock(x0, x1)
y.backward()

print(y.data)
print(x0.grad, x1.grad)
```

```python
x0 = Pedro_Variable(np.array(0.0))
x1 = Pedro_Variable(np.array(2.0))
lr = 1e-3
iters = 10000

for i in range(iters):
    y = rosenbrock(x0, x1)

    x0.cleargrad()
    x1.cleargrad()
    y.backward()

    # using core_simply.py
    # x0.data -= lr * x0.grad
    # x1.data -= lr * x1.grad

    # using core.py
    x0.data -= lr * x0.grad.data
    x1.data -= lr * x1.grad.data

    if (i%100) == 0:
        print(i, x0.data, x1.data, y.data)
```
