```python
np.random.seed(0)
x = Pedro_Variable(np.random.rand(100, 1))
y = 5 + 2*x + np.random.rand(100, 1)

W = Pedro_Variable(np.zeros((1, 1)))
b = Pedro_Variable(np.zeros(1))


lr = 0.1
iters = 100

for i in range(iters):
    y_pred = F.linear_simple(x, W, b)
    loss = F.mean_squared_error(y, y_pred)

    W.cleargrad()
    b.cleargrad()
    loss.backward()

    W.data -= lr * W.grad.data
    b.data -= lr * b.grad.data
    print(W, b, loss)


plt.scatter(x.data, y.data, s=10)
plt.xlabel('x')
plt.ylabel('y')
y_pred = F.linear_simple(x, W, b)
plt.plot(x.data, y_pred.data, color='r')
plt.show()
```

```python
np.random.seed(0)
x = Pedro_Variable(np.random.rand(100, 1))
y = F.sin(2 * np.pi * x) + np.random.rand(100, 1)


I, H, O = 1, 10, 1
W1 = Pedro_Variable(0.01 * np.random.randn(I, H))
b1 = Pedro_Variable(np.zeros(H))
W2 = Pedro_Variable(0.01 * np.random.randn(H, O))
b2 = Pedro_Variable(np.zeros(O))


lr = 0.2
iters = 10000

def predict(x):
    y = F.linear(x, W1, b1)
    y = F.sigmoid(y)
    y = F.linear(y, W2, b2)
    return y


for i in range(iters):
    y_pred = predict(x)
    loss = F.mean_squared_error(y, y_pred)

    W1.cleargrad()
    b1.cleargrad()
    W2.cleargrad()
    b2.cleargrad()
    loss.backward()

    W1.data -= lr * W1.grad.data
    b1.data -= lr * b1.grad.data
    W2.data -= lr * W2.grad.data
    b2.data -= lr * b2.grad.data

    if i % 1000 == 0:
        print(loss)


plt.scatter(x.data, y.data, s=10)
plt.xlabel('x')
plt.ylabel('y')
t = np.arange(0, 1, .01)[:, np.newaxis]
y_pred = predict(t)
plt.plot(t, y_pred.data, color='r')
plt.show()
```
