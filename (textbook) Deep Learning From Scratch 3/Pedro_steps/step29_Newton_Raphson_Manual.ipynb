```python
def f(x):
    y = x**4 - 2 * x**2
    return y

def gx2(x):
    return 12 * x**2 - 4
```

# Newton-Raphson Method

```python
x = Pedro_Variable(np.array(2.0))
iters = 10

for i in range(iters):
    print(i, x)

    y = f(x)
    x.cleargrad()
    y.backward()

    # x.data -= x.grad / gx2(x.data)
    x.data -= x.grad.data / gx2(x.data)
```

# Gradient-Descent Method

```python
x = Pedro_Variable(np.array(2.0))
iters = 100
lr = 0.001

for i in range(iters):
    print(i, x)

    y = f(x)
    x.cleargrad()
    y.backward()

    # x.data -= lr * x.grad
    x.data -= lr * x.grad.data
```
