```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Pedro_Variable:
    def __init__(self, data):
        self.data = data

class Pedro_Function:
    def __call__(self, input):
        x = input.data
        y = self.forward(x)
        output = Pedro_Variable(y)
        return output

    def forward(self, x):
        raise NotImplementedError()

class Pedro_Square(Pedro_Function):
    def forward(self, x):
        return x ** 2

class Pedro_Exp(Pedro_Function):
    def forward(self, x):
        return np.exp(x)
```

```python
def numerical_diff(f, x, eps=1e-4):
    x0 = Pedro_Variable(x.data - eps)
    x1 = Pedro_Variable(x.data + eps)
    y0 = f(x0)
    y1 = f(x1)
    return (y1.data - y0.data) / (2 * eps)
```

```python
f = Pedro_Square()
x = Pedro_Variable(np.array(2.0))
dy = numerical_diff(f, x)
print(dy)
```

```python
def f(x):
    A = Pedro_Square()
    B = Pedro_Exp()
    C = Pedro_Square()
    return C(B(A(x)))

x = Pedro_Variable(np.array(0.5))
dy = numerical_diff(f, x)
print(dy)
```
