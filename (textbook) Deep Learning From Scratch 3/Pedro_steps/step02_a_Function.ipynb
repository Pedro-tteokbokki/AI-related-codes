```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Pedro_Variable:
    def __init__(self, data):
        self.data = data
```

```python
class Pedro_Function:
    def __call__(self, input):
        x = input.data
        y = x ** 2
        output = Pedro_Variable(y)
        return output
```

```python
x = Pedro_Variable(np.array(10))
f = Pedro_Function()
y = f(x)

print(type(y))
print(y.data)
```

```python
class Pedro_Function:
    def __call__(self, input):
        x = input.data
        y = self.forward(x)
        output = Pedro_Variable(y)
        return output

    def forward(self, x):
        raise NotImplementedError()
```

```python
class Pedro_Square(Pedro_Function):
    def forward(self, x):
        return x ** 2
```

```python
x = Pedro_Variable(np.array(10))
f = Pedro_Square()
y = f(x)

print(type(y))
print(y.data)
```
