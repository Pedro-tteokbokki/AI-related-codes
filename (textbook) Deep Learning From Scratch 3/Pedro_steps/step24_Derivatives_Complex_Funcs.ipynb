# 1. Sphere Function
- z = x^2 + y$2

```python
def sphere(x, y):
    z = x**2 + y**2
    return z

x = Pedro_Variable(np.array(1.0))
y = Pedro_Variable(np.array(1.0))
z = sphere(x, y)

z.backward()
print(z.data)
print(x.grad)
print(y.grad)
```

# 2. Matyas Function
- z = 0.26(x^2 + y^2) - 0.48xy

```python
def matyas(x, y):
    z = 0.26 * (x**2 + y**2) - 0.48 * x * y
    return z

x = Pedro_Variable(np.array(1.0))
y = Pedro_Variable(np.array(1.0))
z = matyas(x, y)

z.backward()
print(z.data)
print(x.grad)
print(y.grad)
```
