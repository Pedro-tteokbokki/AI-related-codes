# Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt


def normal(x, mu=0, sigma=1):
    y = 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-(x - mu)**2 / (2 * sigma**2))
    return y
```

```python
x = np.linspace(-5, 5, 100)
y = normal(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)

# mu ====================
y0 = normal(x, mu=-3)
y1 = normal(x, mu=0)
y2 = normal(x, mu=5)

plt.plot(x, y0, label='$\mu$=-3')
plt.plot(x, y1, label='$\mu$=0')
plt.plot(x, y2, label='$\mu$=5')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# sigma ====================
y0 = normal(x, mu=0, sigma=0.5)
y1 = normal(x, mu=0, sigma=1)
y2 = normal(x, mu=0, sigma=2)

plt.plot(x, y0, label='$\sigma$=0.5')
plt.plot(x, y1, label='$\sigma$=1')
plt.plot(x, y2, label='$\sigma$=2')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Central Limit Theorem

```python
N = 3

xs = []
for n in range(N):
    x = np.random.rand()
    xs.append(x)

x_mean = np.mean(xs)
print(x_mean)
```

```python
x_means = []
N = 1

for _ in range(10000):
    xs = []
    for i in range(N):
        x = np.random.rand()
        xs.append(x)
    mean = np.mean(xs)
    x_means.append(mean)

plt.hist(x_means, bins='auto', density=True)
plt.title(f'N={N}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.xlim(-0.05, 1.05)
plt.ylim(0, 5)
plt.show()
```

### Distribution of mean of variables

```python
x_means = []
N = 5

for _ in range(10000):
    xs = []
    for i in range(N):
        x = np.random.rand()
        xs.append(x)
    mean = np.mean(xs)
    x_means.append(mean)

plt.hist(x_means, bins='auto', density=True)
plt.title(f'N={N}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.xlim(-0.05, 1.05)
plt.ylim(0, 5)
plt.show()
```

```python
x_means, x_samples = [], []
N = 5

for _ in range(10000):
    xs = []
    for i in range(N):
        x = np.random.rand()
        xs.append(x)
        x_samples.append(x)
    mean = np.mean(xs)
    x_means.append(mean)

x_norm = np.linspace(-5, 5, 1000)
mu = 1/2
sigma = np.sqrt(1 / 12 / N)
y_norm = normal(x_norm, mu, sigma)


plt.hist(x_means, bins='auto', density=True)
plt.plot(x_norm, y_norm)
plt.title(f'N={N}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.xlim(-0.05, 1.05)
plt.ylim(0, 5)
plt.show()
```

### Distribution of sum of variables

```python
x_sums, x_samples = [], []
N = 5

for _ in range(10000):
    xs = []
    for i in range(N):
        x = np.random.rand()
        xs.append(x)
        x_samples.append(x)
    sums = np.sum(xs)
    x_sums.append(sums)

x_norm = np.linspace(-5, 5, 1000)
mu = 1/2 * N
sigma = np.sqrt(1 / 12 * N)
y_norm = normal(x_norm, mu, sigma)

plt.hist(x_sums, bins='auto', density=True)
plt.plot(x_norm, y_norm)
plt.title(f'N={N}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.xlim(-1, 6)
plt.show()
```
