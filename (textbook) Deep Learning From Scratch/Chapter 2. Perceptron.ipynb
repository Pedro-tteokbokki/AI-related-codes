```python
import numpy as np
```

# AND Gate

```python
def AND1(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
AND1(0, 0), AND1(0, 1), AND1(1, 0), AND1(1, 1)
```

```python
def AND2(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
AND2(0, 0), AND2(0, 1), AND2(1, 0), AND2(1, 1)
```

# NAND Gate

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
NAND(0, 0), NAND(0, 1), NAND(1, 0), NAND(1, 1)
```

# OR Gate

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
OR(0, 0), OR(0, 1), OR(1, 0), OR(1, 1)
```

# XOR Gate

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND2(s1, s2)
    return y
```

```python
XOR(0, 0), XOR(0, 1), XOR(1, 0), XOR(1, 1)
```
